```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = sns.load_dataset('titanic')

print(df.shape)
df.head()
```

```python
#df.isnull().sum()

# Drop missing data column, (deck column as 77% missing data)
df.drop(columns=['deck'],inplace=True)
print(df.shape)
df.columns
```

```python
# Drop redundant columns
df.drop(columns=['alive', 'embark_town', 'class', 'who', 'adult_male'], 
        inplace=True)
print(df.shape)
df.columns
```

```python
#fix missing values
#age
median_age = df['age'].median()
print(f"Median age: {median_age}")

df['age']=df['age'].fillna(median_age) # fillna() replaces NaN values with whatever we pass
print(f"Missing age values: {df['age'].isnull().sum()}")
```

```python
# fill missing values in embarked column with the most common value
print(df['embarked'].value_counts())
print(f"Most common embarked value: {df['embarked'].mode()[0]}")
```

```python
df['embarked']=df['embarked'].fillna(df['embarked'].mode()[0])
print(f"Missing embarked values: {df['embarked'].isnull().sum()}")
```

```python
# Save cleaned dataframe to processed folder
df.to_csv('../data/processed/titanic_cleaned.csv', index=False)
print("Cleaned data saved")
print(f"Shape : {df.shape}")
```
